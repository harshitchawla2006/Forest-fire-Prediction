```python
batters=pd.read_csv("/content/batter.csv")
sample_df=batters.head(100).sample(25,random_state=5)

```

```python
plt.figure(figsize=(12,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
  plt.axhline(125,color='red')
```

```python
# second way of plotting graphs
fig,ax = plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'],batters['strike_rate'],color='blue')
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')
# with this we can plot multiple plots on same axis
```

```python
# 2 scatter plots side by side one (avg v/s strike rate) and (avg vs runs)
fig,ax=plt.subplots(nrows=2,ncols=1,figsize=(12,6),sharex=True)
ax[0].scatter(batters['avg'],batters['strike_rate'],color='blue',s=10)
ax[1].scatter(batters['avg'],batters['runs'],color='red',s=7)
ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')
#sharex=True → all subplots use the same x-axis limits and ticks.

ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
# 3rd way
fig=plt.figure()
ax1=fig.add_subplot(3,1,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='blue',s=10)
ax2=fig.add_subplot(3,1,2)
ax2.scatter(batters['avg'],batters['runs'],color='red',s=7)
ax3=fig.add_subplot(3,1,3)
ax3.hist(batters['runs'],color='green')
```

```python
# 3D scatter plots
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(batters['avg'],batters['strike_rate'],batters['runs'])
```

```python
# 3D line plot
x = [0,1,5,25]
y = [0,10,13,0]
z = [0,13,20,9]
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z,color='red')
```

```python
# surface plot
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
z=xx**2+yy**2
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
```
