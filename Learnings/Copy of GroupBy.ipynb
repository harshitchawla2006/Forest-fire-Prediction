```python
# Group Attributes and Methods
# len(movies.groupby('Genre'))
#or
# movies['Genre'].nunique()
genres=movies.groupby('Genre')
#genres.get_group('Fantasy') # or movies[movies['Genre']=='Fantasy'] but first is more fast and efficent
# genres.groups
movies
```

```python
# agg
genres=movies.groupby('Genre')
genres.agg(
    {
        'Runtime':'mean',
        'Gross':'sum',
        'IMDB_Rating':'mean'


    }
)
```

```python
# looping on groups
# for group,data in genres:
#     print(group)  # all genres
#     # print(data) # dataframes
 # highest rated movie of each genre using loops
for group,data in genres:

    print(data[data['IMDB_Rating']==data['IMDB_Rating'].max()]['Series_Title'])
```

```python
# apply
genres.apply(min)
```

```python
# find number of movies of each genre starting with A
def foo(group):
  return group['Series_Title'].str.startswith('A').sum()

genres.apply(foo)

```

```python
# find ranking of each movie in the group acc to IMDB_rating
def rank_movie(group):
  group['Genre_rank']=group['IMDB_Rating'].rank(ascending=False)
  return group

genres.apply(rank_movie)
```

```python
# get normalized rating for every movie
def normalize_rating(group):
    min_val = group['IMDB_Rating'].min()
    max_val = group['IMDB_Rating'].max()
    if max_val == min_val:
        group['normalized_rating'] = 0  # or 1, or NaN depending on what you want
    else:
        group['normalized_rating'] = (group['IMDB_Rating'] - min_val) / (max_val - min_val)
    return group

genres.apply(normalize_rating)
```

```python
ipl=pd.read_csv("deliveries.csv")
ipl
```

```python
# find top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with most number of sixes
ipl[ipl['batsman_runs']==6].groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
# find batsman with most number of 4's or 6's in last 5 overs
temp_df=ipl[ipl['over']>15]
temp_df=temp_df[(temp_df['batsman_runs']==4)|(temp_df['batsman_runs']==6)]
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# find V Kholi's record against all teams
temp_df=ipl[ipl['batsman']=='V Kohli']

new_df=temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
new_df
```

```python
# create a function that return maxm score of any batsman
def highest_score(batsman):
  temp_df=ipl[ipl['batsman']==batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]

highest_score('CH Gayle')
```
